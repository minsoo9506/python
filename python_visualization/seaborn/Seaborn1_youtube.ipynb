# PairGrid

```python
x = sns.PairGrid(iris)
x = x.map(plt.scatter)
```

```python
x = sns.PairGrid(iris)
x = x.map_diag(plt.hist) #diag만 하면 diag만 나옴
```

```python
x = sns.PairGrid(iris, hue = 'species')
x = x.map_diag(plt.hist)
x = x.map_offdiag(plt.scatter)
x = x.add_legend() #해당 hue가 어떤건지 보여줌
```

```python
x = sns.PairGrid(iris, hue = 'species')
x = x.map_diag(plt.hist, histtype = 'step', linewidth = 4) # histtype으로 hist조정
x = x.map_offdiag(plt.scatter)
x = x.add_legend()
```

```python
x = sns.PairGrid(iris, vars= ['petal_length', 'petal_width'], hue ='species')
x = x.map_diag(plt.hist, edgecolor = 'black') # vars 로 원하는 변수만 고를 수 있음
x = x.map_offdiag(plt.scatter, edgecolor = 'black') #edgecolor로 그림처럼 표현
x = x.add_legend()
```

```python
x = sns.PairGrid(iris , x_vars = ['petal_length', 'petal_width'],
                y_vars = ['sepal_length', 'sepal_width'] )
x = x.map(plt.scatter)
```

```python
x = sns.PairGrid(iris, hue = 'species')
x = x.map_diag(plt.hist)
x = x.map_upper(plt.scatter) #lower 와 upper로 원하는 대로
x = x.map_lower(sns.kdeplot)
x = x.add_legend()
```

```python
x = sns.PairGrid(iris, hue = 'species', hue_kws = {'marker':['D','s','+']})
x = x.map(plt.scatter,s=30, edgecolor = 'black')
x = x.add_legend()
#hue_kws : marker를 원하는 모양으로!
```
